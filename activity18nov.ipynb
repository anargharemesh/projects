```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
```

```python
data=pd.read_csv('Telco-Customer-Churn.xlsx - Telco-Customer-Churn.csv')
```

```python
data.head()
```

# 1 Compare churn count with respect to gender. 

```python
plt.figure(figsize=(8,5))
data1=data[data['Churn']=='Yes' ]

sns.countplot(x=data1['Churn'],hue=data1['gender'])
plt.grid()
```

# 2) Find out how many female senior citizens there in the dataset 

```python
fem_senior=data[(data['SeniorCitizen']==1) & (data['gender']=='Female')]
fem_senior['SeniorCitizen'].value_counts()
```

```python
sns.countplot(x=fem_senior['SeniorCitizen'])
plt.xlabel("Female senior-citizen")
```

# 3) Compare 'tenure' with 'Total Charges'
