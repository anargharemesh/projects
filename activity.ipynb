```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
```

```python
data=pd.read_csv('diabetes_model.csv')
```

```python
data.head()
```

```python
sns.countplot(data['Outcome'])
warnings.filterwarnings("ignore")
plt.grid()
```

```python

ddata=pd.DataFrame(data)
print(ddata)
sns.boxplot(x=ddata['Outcome'],y=ddata['Age'],data=ddata)
plt.title('age v/s outcome')
plt.show()
```

```python
sns.heatmap(ddata.corr(),annot=True)
plt.xticks(rotation=45)
```

```python
sns.violinplot(x=ddata['Outcome'],y=ddata['DiabetesPedigreeFunction'],data=ddata)
```

```python
plt.figure(figsize=(10,10))
sns.scatterplot(x=ddata['Insulin'],y=ddata['Glucose'],data=ddata,hue=ddata['Outcome'])
```
