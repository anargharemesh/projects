```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
```

```python
data=pd.read_excel('Dataset_w6.xlsx')
data.head()
```

# 1. Find out how many countries are there in the dataset

```python
data['country'].value_counts()
```

```python
data['country'].count()
```

# 2. Examine the data by using visualizations.

```python
plt.figure(figsize=(12,10))
sns.barplot(x='Region',y='Population in thousands (2017)',data=data)
plt.xticks(rotation=90)
```

```python
plt.figure(figsize=(12,10))
sns.violinplot(x='Region',y='Surface area (km2)',data=data)
plt.xticks(rotation=90)
```

```python
sns.boxplot()
```

```python
plt.figure(figsize=(12,10))
sns.boxplot(x='Region',y='International trade: Imports (million US$)',data=data)
plt.xticks(rotation=90)
```

# 3. Find out how many countries with respect to the region are availablein the dataset

```python
data.groupby('Region')['country'].count() 
```

# 4. On average which region has the highest
a. GDP per capita (current US$)
b. International trade: Imports (million US$)

```python
#region with highest of International trade: Imports (million US$) on average
data.groupby('Region')['International trade: Imports (million US$)'].mean().nlargest(1)
```

```python
#region with highest of GDP per capita (current US ) on average
data.groupby('Region')['GDP per capita (current US$)'].mean().nlargest(1)
```

# 5. Find out which region has more consistent with respect to GDP per capita (current US$)

```python
#least variance value is the factor for finding consistent region  with respect to GDP
data.groupby('Region')['GDP per capita (current US$)'].var().nsmallest(1)
```
